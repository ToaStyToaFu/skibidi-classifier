```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

demo_dir = {
    'demo1': 'demo/demo.png',
    'demo2': 'demo/demo2.jpg',
    'demo3': 'demo/demo3.jpg',
    'demo4': 'demo/demo4.png',
    'demo5': 'demo/demo5.jpg'
}

model = load_model('model.keras')

def classify(dir, real_class):
    img = mpimg.imread(dir)
    plt.title(f'Real class: {real_class}')
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    input = image.load_img(dir, target_size=(256,256), color_mode='grayscale')
    input_arr = image.img_to_array(input)
    input_arr = np.expand_dims(input_arr, axis =0)
    input_arr /= 255

    pred = model.predict(input_arr)

    print(f"Predicted Class: {'Human' if pred > 0.5 else 'AI'}")
    print(pred)
```

```python
classify(demo_dir['demo1'], 'AI')
classify(demo_dir['demo2'], 'AI')
classify(demo_dir['demo3'], 'Human')
classify(demo_dir['demo4'], 'Human')
classify(demo_dir['demo5'], 'Human')
```
